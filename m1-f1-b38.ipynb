# Empirical bias–variance decomposition at a single point


```python
import numpy as np

rng = np.random.default_rng(42)
n_trials = 10_000
n_samples = 20
noise_std = 0.1
x0 = 0.5
f_true_x0 = np.sin(np.pi * x0)  # = 1.0

pred_linear = np.empty(n_trials)
pred_poly15 = np.empty(n_trials)

for t in range(n_trials):
    x_train = rng.uniform(-1, 1, n_samples)
    y_train = np.sin(np.pi * x_train) + rng.normal(0, noise_std, n_samples)

    # Straight line
    c_lin = np.polyfit(x_train, y_train, deg=1)
    pred_linear[t] = np.polyval(c_lin, x0)

    # Degree-15 polynomial
    c_p15 = np.polyfit(x_train, y_train, deg=15)
    pred_poly15[t] = np.polyval(c_p15, x0)

def report(name, preds):
    mean_pred = preds.mean()
    bias_sq   = (mean_pred - f_true_x0) ** 2
    variance  = preds.var()
    total     = bias_sq + variance + noise_std ** 2
    print(f"{name:12s}  mean={mean_pred:+.4f}  "
          f"bias²={bias_sq:.4f}  var={variance:.4f}  "
          f"noise={noise_std**2:.4f}  total={total:.4f}")

print(f"True f(x0) = {f_true_x0:.4f}")
report("Linear",     pred_linear)
report("Poly deg-15", pred_poly15)
```
