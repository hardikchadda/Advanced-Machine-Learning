# Empirical bias–variance decomposition at a single point

_From the **Advanced Machine Learning** course on Zylo._

Run all cells (`Runtime → Run all` or `Ctrl+F9`) to execute.

```python
import numpy as np

# ── Setup ─────────────────────────────────────────────────────────
rng = np.random.default_rng(42)
n_trials = 10_000      # number of fresh training sets
n_samples = 20         # size of each training set
noise_std = 0.1
x0 = 0.5               # the test point we evaluate at
f_true_x0 = np.sin(np.pi * x0)  # = 1.0

# ── Storage for the 10,000 predictions from each model ──────────
pred_linear = np.empty(n_trials)
pred_poly5  = np.empty(n_trials)

# ── The experiment ──────────────────────────────────────────────
for t in range(n_trials):
    # Draw a fresh training set
    x_train = rng.uniform(-1, 1, n_samples)
    y_train = np.sin(np.pi * x_train) + rng.normal(0, noise_std, n_samples)

    # Fit a linear model:  polyfit with degree 1
    coeffs_lin = np.polyfit(x_train, y_train, deg=1)
    pred_linear[t] = np.polyval(coeffs_lin, x0)

    # Fit a degree-5 polynomial
    coeffs_p5 = np.polyfit(x_train, y_train, deg=5)
    pred_poly5[t] = np.polyval(coeffs_p5, x0)

# ── Estimate bias and variance for each model ──────────────────
def bv_report(name, preds, f_true):
    mean_pred = preds.mean()
    bias_sq   = (mean_pred - f_true) ** 2
    variance  = preds.var()
    total     = bias_sq + variance + noise_std ** 2
    print(f"{name:12s}  mean={mean_pred:.4f}  "
          f"bias²={bias_sq:.4f}  var={variance:.4f}  "
          f"noise={noise_std**2:.4f}  total MSE={total:.4f}")

print(f"True f(x0) = {f_true_x0:.4f}")
bv_report("Linear",    pred_linear, f_true_x0)
bv_report("Poly deg-5", pred_poly5,  f_true_x0)
```
